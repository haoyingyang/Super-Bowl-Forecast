```python
import pandas as pd
from pandas import ExcelWriter
from pandas import ExcelFile
 
df = pd.read_excel('Desktop/NFL 2018 Season Data.xlsx', sheet_name='Sheet1')
import numpy as np

df['wins']= np.where(df.net_score>0,1,0)
df
```

```python
import numpy as np

df['wins']= np.where(df.net_score>0,1,0)
```

```python
team_stat=df.groupby('Team').mean()
team_stat=team_stat.drop('Game Number',1)
```

```python
team_stat.rename(columns={'wins': 'win_rate', 'net_score':'avg_ns'}, inplace=True)
team_stat['Team_Rank'] = team_stat['win_rate'].rank(ascending=False,method='min')
team_stat=team_stat.reset_index()
team_stat
```

```python
Regression=team_stat[['Team','Team_Score','Team_Passing','Team_Rushing','Team_Turnovers','Team_total_yards','win_rate']]
Team=df[['Team',"Opponent",'Team_Score','Opponent_Score','net_score']]
Regression.rename(columns={'Team': 'Opponent',"Team_Score":"Opp_avg_Score","Team_Passing":"Opp_avg_Passing","Team_Turnovers":"Opp_avg_to",
                                     'Team_total_yards':'Opp_avg_yards','win_rate':'Opp_win_rate'}, inplace=True)
Reg=pd.merge(Team, Regression,how='left', left_index=False, right_index=False, copy=False, indicator=False)
Reg
```

```python
Rg=Reg.loc[(Reg['Team'] == 'New England Patriots') | (Reg['Team'] =='Los Angeles Rams')]
Rg.to_csv('/Users/yanghaoying/Desktop/Rg.csv')
```
